```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8m.pt")
```

```python
results = model.predict("she.jpg")
```

```python
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
 
plt.title("She Image")
plt.xlabel("X pixel scaling")
plt.ylabel("Y pixels scaling")
 
image = mpimg.imread("she.jpg")
plt.imshow(image)
```

```python
result = results[0]
```

```python
len(result.boxes)
```

```python
box = result.boxes[0] #we are taiking first box here
```

```python
#print the details in first box
print("Object type:", box.cls) 
print("Coordinates:", box.xyxy)
print("Probability:", box.conf)
```

```python
cords = box.xyxy[0].tolist()
class_id = box.cls[0].item()
conf = box.conf[0].item()
print("Object type:", class_id)
print("Coordinates:", cords)
print("Probability:", conf)
```

```python
print(result.names)
```

```python
#here the object type is 0 so we need to find which object is that from result.names
```

```python
cords = box.xyxy[0].tolist()
cords = [round(x) for x in cords]
class_id = result.names[box.cls[0].item()]
conf = round(box.conf[0].item(), 2)
print("Object type:", class_id)
print("Coordinates:", cords)
print("Probability:", conf)
```
